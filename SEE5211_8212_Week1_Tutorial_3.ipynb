### Pie chart

```python
# Frequencies
import matplotlib.pyplot as plt # imports a plotting library
sizes = [60, 54, 21, 10, 5]

# Labels for each section
labels = ['1', '2', '3', '4', '5']

# Plot the pie chart
plt.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%', 
    shadow=True, 
    startangle=90,
    pctdistance=1.7,  
    labeldistance=1.1,  
    textprops={'fontsize': 20} 
)


# Show the plot
plt.show()
```

```python
### Histogram
```

```python
# Given data set
data = [
    6.9, 7.8, 8.9, 5.2, 7.7, 9.6, 8.7,
    6.7, 4.8, 8.0, 10.1, 8.5, 6.5,
    9.2, 7.4, 6.3, 5.6, 7.3, 8.3, 7.2,
    7.5, 6.1, 9.4, 5.4, 7.6, 8.1, 7.9
]

# Create the histogram
plt.hist(data, bins='auto', edgecolor='black', alpha=0.7)

plt.title('Histogram of Data', fontsize=20)
plt.xlabel('Nitrate (mg/L)', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Show the histogram
plt.show()
```

```python
# Create the histogram with relative frequencies
plt.hist(data, bins='auto', edgecolor='black',
         alpha=0.7, density=True)

plt.title('Relative Frequency Histogram', fontsize=20)
plt.xlabel('Nitrate (mg/L)', fontsize=20)
plt.ylabel('Relative Frequency', fontsize=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Show the histogram
plt.show()
```

### Cumulative relative frequency

```python

# Rainfall categories (in inches)
rainfall_categories = ['4 to <5', '5 to <6', '6 to <7', '7 to <8', '8 to <9', 
                       '9 to <10', '10 to <11', '11 to <12', '12 to <13', '13 to <14']

# Midpoints of each rainfall category for plotting purposes
midpoints = [4.5, 5.5, 6.5, 7.5, 8.5, 9.5, 10.5, 11.5, 12.5, 13.5]

# Relative frequencies
relative_frequencies = [0.052, 0.103, 0.086, 0.103, 0.172, 0.069, 0.207, 0.103, 0.052, 0.052]
```

```python
# Calculate the cumulative relative frequency
cumulative_relative_frequencies = [sum(relative_frequencies[:i+1]) for i in range(len(relative_frequencies))]

# Create a dot plot for the cumulative relative frequency
plt.plot(midpoints, cumulative_relative_frequencies, 'o-')

plt.xlabel('Annual Rainfall (in inches)', fontsize=20)
plt.ylabel('Cumulative Relative Frequency', fontsize=20)

# Display the plot
plt.show()
```

### Scatter plot

```python
heights = [160, 160, 170, 175, 170, 185, 180, 190, 200, 205, 150, 155,170,190]  # Heights in centimeters
weights = [50, 55, 60, 65, 75, 80, 85, 90, 95, 100, 45, 50,65,80]              # Weights in kilograms
```

### Boxplot

```python
# Multiple sets of example data
data1 = [20, 21, 22, 23, 24, 25, 26, 27, 28]
data2 = [25, 26, 27, 28, 29, 30, 31, 32, 33]
data3 = [30, 31, 32, 33, 34, 35, 36, 37, 38]

# Create side-by-side box plots
plt.boxplot([data1, data2, data3])

# Set the title and labels
plt.title('Multiple Box Plots Example', fontsize=20)
plt.ylabel('Values', fontsize=16)
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3']) 
plt.show()
```
